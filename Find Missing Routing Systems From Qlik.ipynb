```python
import pandas as pd
import numpy as np
```

```python
#Reading in all data from Qliksense

qlik_addresses = pd.read_excel("qlik address.xlsx")
qlik_addresses
```

```python
#finding all rows where there is no routing system listed

missing_routing = qlik_addresses[qlik_addresses["TTY Routing System"] == "-"]
missing_routing
```

```python
#drop duplicate destination addresses

missing_routing_singular = missing_routing.drop_duplicates(subset = "Destination Address")
```

```python
#exporting it to Excel

missing_routing_singular.to_excel("Destination Addresses With No Routing.xlsx")
```
